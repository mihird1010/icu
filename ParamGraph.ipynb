```python
from datetime import datetime
import pandas as pd
#%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('132539.txt')
```

```python
df.head()
```

```python
arrt=[]
arrv=[]
for index,row in df.iterrows():
    if row['Parameter']=='HR':
        arrt.append(row['Time'])
        arrv.append(row['Value'])
```

```python

#plt.plot(arrt,arrv)
#plt.gcf().autofmt_xdate()

index = [i for i in range(len(arrt))]
time = arrt
plt.figure(figsize=(20,7))
plt.plot(index, arrv)
plt.xticks(index, time)
plt.gcf().autofmt_xdate()
plt.show()
```
