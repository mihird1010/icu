```python
from datetime import datetime
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def findDelta(vals):
    
    delta={}
    for i in vals:
        #print(i)
        if len(vals[i])>1:
            tot=0
            for j in range(len(vals[i])-1):
                #print(j)
                d=abs(vals[i][j+1]-vals[i][j])
                if d!=0 and vals[i][j]!=0:
                    tot=tot+d/vals[i][j]
                    delta[i]=tot
                elif d!=0 and vals[i][j+1]!=0:
                    tot=tot+d/vals[i][j+1]
                    delta[i]=tot
                else:
                    delta[i]=0
        else:
            delta[i]=0
    sum=0
    for i in delta:
        sum=sum+delta[i]
    return sum
```

```python
def findNorm(vals):
    normvals={'Albumin':4.5,'ALP':95,'ALT':30,'AST':25,'Bilirubin':0.1,'BUN':15,'Cholesterol':50,'Creatinine':0.9,'DiasABP':80,'FiO2':0.3,'GCS':15,'Glucose':80,'HCO3':25,'HCT':40,'HR':72,'K':4.2,'Lactate':1,'Mg':2,'MAP':95,'Na':140,'NIDiasABP':80,'NIMAP':95,'NISysABP':120,'PaCO2':40,'PaO2':200,'pH':7.4,'Platelets':300,'RespRate':18,'SaO2':97,'SysABP':120,'Temp':37,'TroponinT':1.2,'WBC':10}
    normdel={}
    for i in normvals:
        if len(vals[i])>=1:
            tot=0
            for j in range(len(vals[i])):
                #print(j)
                tot=tot+(abs(normvals[i]-vals[i][j]))/normvals[i]
            normdel[i]=tot
        else:
            normdel[i]=0
    
    sum=0
    for i in normdel:
        sum=sum+normdel[i]
    return sum
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN()
```

```python
dbscan.fit(patients)
```

```python
dbscan.labels_
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2).fit(patients)
pca_2d =pca.transform(patients)
```

```python
for i in range(0, pca_2d.shape[0]):
    if dbscan.labels_[i] == -1:
        c1 = plt.scatter(pca_2d[i,0],pca_2d[i,1],c='r')
    elif dbscan.labels_[i] == 1:
        c2 = plt.scatter(pca_2d[i,0],pca_2d[i,1],c='g')
    elif dbscan.labels_[i] == 0:
        c3 = plt.scatter(pca_2d[i,0],pca_2d[i,1],c='b')
plt.legend([c1, c2, c3], ['Cluster 1', 'Cluster 2','Noise'])
plt.title('DBSCAN finds 2 clusters and noise')
```
