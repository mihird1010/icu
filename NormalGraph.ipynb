```python
normdel={}
for i in normvals:
    if len(vals[i])>=1:
        tot=0
        for j in range(len(vals[i])):
            #print(j)
            tot=tot+(abs(normvals[i]-vals[i][j]))/normvals[i]
        normdel[i]=tot
    else:
        normdel[i]=0
```
